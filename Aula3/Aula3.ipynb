Import all libs and set the browser webdriver path:

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

driver_path = "C:/Users/mathe/intensivo-hashtag/Aula3/chromedriver.exe"
brave_path = "C:/Program Files/BraveSoftware/Brave-Browser/Application/brave.exe"

option = webdriver.ChromeOptions()
option.binary_location = brave_path
# option.add_argument("--incognito") OPTIONAL
# option.add_argument("--headless") OPTIONAL

# Create new Instance of Chrome
browser = webdriver.Chrome(executable_path=driver_path, options=option)
```

Get dollar price in reais:

```python
browser.get("https://google.com")
browser.find_element_by_xpath(
    '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("dólar")

browser.find_element_by_xpath(
    '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)

dollar_price = browser.find_element_by_xpath(
    '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute("data-value")

print(dollar_price)
```

Get euro price in reais:

```python
browser.get("https://google.com")
browser.find_element_by_xpath(
    '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("euro")

browser.find_element_by_xpath(
    '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)

euro_price = browser.find_element_by_xpath(
    '//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute("data-value")

print(euro_price)
```

Get gold price in reais:

```python
browser.get("https://www.melhorcambio.com/ouro-hoje")

gold_price = browser.find_element_by_xpath(
    '//*[@id="comercial"]').get_attribute("value")

#Replace comma with dot
gold_price = gold_price.replace(",", ".")

print(gold_price)
```

Read "Produtos.xlsx" and display:

```python
spreadsheet = pd.read_excel("Produtos.xlsx")
display(spreadsheet)
```

Update currencies prices:

```python
spreadsheet.loc[spreadsheet["Moeda"] == "Dólar", "Cotação"] = float(dollar_price)
spreadsheet.loc[spreadsheet["Moeda"] == "Euro", "Cotação"] = float(euro_price)
spreadsheet.loc[spreadsheet["Moeda"] == "Ouro", "Cotação"] = float(gold_price)

display(spreadsheet)
```

Update converted prices and final prices:

```python
spreadsheet["Preço de Base Reais"] = spreadsheet["Preço Base Original"] * spreadsheet["Cotação"]
spreadsheet["Preço Final"] = spreadsheet["Preço de Base Reais"] * spreadsheet["Margem"]
display(spreadsheet)
```

Create new excel spreadsheet with updated values and end browser session:

```python
spreadsheet.to_excel("Produtos Novo.xlsx", index=False)
browser.quit()
```
